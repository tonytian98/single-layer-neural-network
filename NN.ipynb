```python
import numpy as np
import pandas as pd
import math as mt
```

```python
def b(v,x,r):
    v=v.reshape(h,p)
    x=x.reshape(p,1)
    r=r.reshape(h,1)
    return 1/(mt.e**(-np.matmul(v,x)+r)+1)
def output(b,w,o):
    w=w.reshape(1,h)
    b=b.reshape(h,1)
    return 1/(mt.e**(-np.matmul(w,b)+o)+1)
###mean-square error
def g_1(y,yhat):
    return yhat*(1-yhat)*(y-yhat)
###Cross entropy
def g_2(y,yhat):
    return y+yhat-2*y*yhat
```

```python
df=pd.read_csv(r'C:\Users\Admin\Downloads\horseColicTraining2.txt',sep='\t',header=None)
X=df.drop(columns=21)
Y=df[[21]]
```

```python
def f(x):
    if x<=0.5:
        return 0
    if x>0.5:
        return 1
```

```python
names=['mean-square error','Cross entropy']
namei=0
for fff in [g_1,g_2]:
    
    ### h is the number of hidden units in the layer（8），c is learning rate（0.1）
    h=8
    p=X.iloc[0].shape[0]
    n=len(Y.index)
    c=0.1
    ###all initial numbers are 0.01
    v=np.ones((h,p))*0.01
    r=np.ones(h).reshape(h,1)*0.01
    o=np.ones(1)*0.01
    w=np.ones(h).reshape(h,1)*0.01
    ###iterate 400 times
    for it in range(400):
        for i in range(len(X.index)):
            x=X.iloc[i].to_numpy()
            bi=b(v,x,r)
            yhat=output(bi,w,o)[0,0]
            y=Y.iloc[i].values[0]
            g=fff(y,yhat)
            e=bi*(1-bi)*w*g
            x=x.reshape(1,21)
            w=w+c*g*bi
            o=o-c*g
            v=v+c*np.matmul(e,x)
            r=r-c*e
    score=0
    for i in range(len(X.index)):
        x=X.iloc[i].to_numpy()
        bi=b(v,x,r)
        score+=abs(Y.iloc[i].values[0]-f(output(bi,w,o)[0,0]))
    print(names[namei],':\nscore：',1-score/n)
    print(w)
    print(v)
    print(r)
    print(o)
    namei+=1
```
